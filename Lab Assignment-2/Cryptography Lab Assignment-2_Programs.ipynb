```python
# In the below function r1==a and r2==b
def extended_eculedian_algo(r1,r2):
    a,b=r1,r2
    # Below two lines are used for declaring the values of s1,s2,t1,t2
    s1=t2=1
    s2=t1=0
    # The below while loop runs until r2!=0
    while r2!=0:
        # Finding quotient and remainder
        q=r1//r2
        r=r1%r2
        # Assigning the values of r1,r2,s1,s2,t1,t2
        # s=s1-(q*s2) and t=t1-(q*t2)
        r1,r2,s1,s2,t1,t2 = r2,r,s2,s1-(q*s2),t2,t1-(q*t2)
        # Calculation of gcd and returning the value
    final_ans=(s1*a)+(t1*b)
    return(final_ans)
        
# Below Two lines take the input from user   
a=int(input("Enter the value of a:"))
b=int(input("Enter the value of b:"))
# Assigning the function to a variable called 'res'
res=extended_eculedian_algo(a,b)
#printing the final result
print("The GCD is:",res)
```
